```python
import time
import pandas as pd
import requests
from bs4 import BeautifulSoup
from csv import reader

urls = []
data = []

rating_mapping = {
    'One': 1,
    'Two': 2,
    'Three': 3,
    'Four': 4,
    'Five': 5
}

with open('bookslinks.csv', 'r') as f:
    csv_reader = reader(f)
    for row in csv_reader:
        urls.append(row[0])

def transform(url):
    r = requests.get(str(url))
    soup = BeautifulSoup(r.content, 'html.parser')
    
    title = soup.find('h1').text
    star_rating = soup.find('p', class_='star-rating')['class'][-1]
    star_rating = rating_mapping.get(star_rating)
    
    product_description = None
    try:
        h2_tag = soup.find('h2', text='Product Description')
        if h2_tag:
            p_tag = h2_tag.find_next('p')
            if p_tag:
                product_description = p_tag.text
    except AttributeError:
        pass
    
    data.append({'Title': title, 'StarRating': star_rating, 'ProductDescription': product_description})
    print(title)

start_time = time.time()

for url in urls:
    transform(url)

end_time = time.time()
duration = end_time - start_time

print(f"Total titles: {len(data)}")
print(f"Duration: {duration} seconds")

df = pd.DataFrame(data)
df.to_csv('book_data.csv', index=False)
print('Complete.')
```

```python
import pandas as pd
import numpy as np
df = pd.read_csv('book_data_.csv') 
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df = df[df['StarRating'] != 3]

df['Positively Rated'] = np.where(df['StarRating'] > 3, 1, 0)
```

```python
df.head(10)
```

```python
# Büyük-küçük dönüşümü
df['ProductDescription'] = df['ProductDescription'].apply(lambda x: " ".join(x.lower() for x in x.split()))

# Noktalama işaretleri
df['ProductDescription'] = df['ProductDescription'].str.replace('[^\w\s]','')

# Sayılar
df['ProductDescription'] = df['ProductDescription'].str.replace('\d','')

# Stopwords
import nltk
from nltk.corpus import stopwords

# nltk.download('stopwords')


# Seyreklerin silinmesi
sil = pd.Series(' '.join(df['ProductDescription']).split()).value_counts()[-1000:]
df['ProductDescription'] = df['ProductDescription'].apply(lambda x: " ".join(x for x in x.split() if x not in sil))

# Lemmatization


```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df['ProductDescription'], 
                                                    df['Positively Rated'], 
                                                    random_state=0)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score

# CountVectorizer nesnesini oluşturma ve verileri dönüştürme
vect = CountVectorizer().fit(X_train)

# Dönüştürülen verilerin feature isimlerini alma
feature_names = vect.get_feature_names_out()

# Her 500. feature ismini yazdırma
print(feature_names[::500])

# Verileri dönüştürme
X_train_vectorized = vect.transform(X_train)

# Modeli oluşturma ve eğitme
model = LogisticRegression()
model.fit(X_train_vectorized, y_train)

# Tahmin yapma
tahmin = model.predict(vect.transform(X_test))

# AUC hesaplama
print('AUC: ', roc_auc_score(y_test, tahmin))

# Katsayıları sıralama
sıralanmıs = model.coef_[0].argsort()
```

```python
print('olumsuz :\n{}\n'.format(feature_names[sıralanmıs[:10]]))
print('olumlu: \n{}'.format(feature_names[sıralanmıs[:-11:-1]]))
```
